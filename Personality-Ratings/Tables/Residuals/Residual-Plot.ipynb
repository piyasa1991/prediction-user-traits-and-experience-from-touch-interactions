```python
#Neuroticism

# best models
xl_best1 = pd.read_excel('feature_selection_alltargets_mahalanobis_PCA_0.95_predictions.xlsx',sheet_name='MARS')
xl_best2 = pd.read_excel('feature_selection_alltargets_mahalanobis_PCA_0.8_predictions.xlsx',
                      sheet_name='MARS')

# worst models
xl_worst1 = pd.read_excel('feature_selection_alltargets_mahalanobis_PCA_0.95_predictions.xlsx',sheet_name='Lasso Regression')
xl_worst2 = pd.read_excel('feature_selection_alltargets_mahalanobis_transformed_PCA_0.95_predictions.xlsx',sheet_name='Lasso Regression')

fig ,ax = plt.subplots(2,2,figsize=(20,8),sharey=True)
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

# best models
xl_best1.hist(column='Residuals_Neuroticism', bins=12, grid=True, color='#86bf91', zorder=2, rwidth=0.9, ax=ax[0][0])
stats, p = kstest(xl_best1['Residuals_Neuroticism'],'norm')
mean = xl_best1['Residuals_Neuroticism'].mean()
sd = xl_best1['Residuals_Neuroticism'].std()
# place text
ax[0][0].text(1.5,max(ax[0][0].get_yticks())-10,'Stats:%0.2f\n p:%0.2f\nNormal:%s \nMean:%0.2f \nSD:%0.2f'%(stats,p,p>0.05,mean,sd),
                           ha="center", va="center", color="black",bbox=props)
ax[0][0].set_title('Residuals of Neuroticism in best model (MARS)')

stats, p = kstest(xl_best2['Residuals_Neuroticism'],'norm')
mean = xl_best2['Residuals_Neuroticism'].mean()
sd = xl_best2['Residuals_Neuroticism'].std()
xl_best2.hist(column='Residuals_Neuroticism', bins=12, grid=True, color='#86bf91', zorder=2, rwidth=0.9, ax=ax[0][1])
ax[0][1].text(1.5,max(ax[0][1].get_yticks())-10,'Stats:%0.2f\n p:%0.2f\nNormal:%s \nMean:%0.2f \nSD:%0.2f'%(stats,p,p>0.05,mean,sd),
                           ha="center", va="center", color="black",bbox=props)
ax[0][1].set_title('Residuals of Neuroticism in second best model (MARS)')



# worst models
stats, p = kstest(xl_worst1['Residuals_Neuroticism'],'norm')
mean = xl_worst1['Residuals_Neuroticism'].mean()
sd = xl_worst1['Residuals_Neuroticism'].std()
xl_worst1.hist(column='Residuals_Neuroticism', bins=12, grid=True, color='#86bf91', zorder=2, rwidth=0.9, ax=ax[1][0])
ax[1][0].text(1.5,max(ax[1][0].get_yticks())-10,'Stats:%0.2f\n p:%0.2f\nNormal:%s\nMean:%0.2f \nSD:%0.2f'%(stats,p,p>0.05,mean,sd),
                           ha="center", va="center", color="black",bbox=props)
xl_worst2.hist(column='Residuals_Neuroticism', bins=12, grid=True, color='#86bf91', zorder=2, rwidth=0.9, ax=ax[1][1])
stats, p = kstest(xl_worst2['Residuals_Neuroticism'],'norm')
mean = xl_worst2['Residuals_Neuroticism'].mean()
sd = xl_worst2['Residuals_Neuroticism'].std()
ax[1][1].text(1.5,max(ax[1][1].get_yticks())-10,'Stats:%0.2f\n p:%0.2f\nNormal:%s\nMean:%0.2f \nSD:%0.2f'%(stats,p,p>0.05,mean,sd),
                           ha="center", va="center", color="black",bbox=props)

# set_style(ax)
ax[1][0].set_title('Residuals of Neuroticism in worst model (Lasso)')
ax[1][1].set_title('Residuals of Neuroticism in second worst model (Lasso)')

fig.suptitle("Residuals for Neuroticism of best and worst models\n upper is for best and lower is for worst",weight='bold',fontsize=15)
```
