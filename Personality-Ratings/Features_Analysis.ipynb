# Analyses the behavior/characteristic/distributions for the features

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
'''Analysis of the descriptors based on users'''
df_descriptors=pd.read_csv('../datasets/files_generated/study1_cleandata.csv')
# print(df_descriptors.columns)
df2=df_descriptors[df_descriptors['user_id']==100]
with PdfPages('Figures/study1_descriptors_distribution_spell_user_id.pdf') as pdf:
    df2 = df2.select_dtypes('float64')
    # df2=df2[df2['sessionNr']==1]
    l = df2.columns.values
    number_of_columns=12
    number_of_rows = len(l)-1/number_of_columns
    plt.figure(figsize=(2*number_of_columns,5*number_of_rows))
    for i in range(0,len(l)):
        fig=plt.subplot(number_of_rows + 1,4,i+1)
        sns.distplot(df2[l[i]],kde=True,hist=True)
    pdf.savefig(bbox_inches='tight')
```

```python
df = pd.read_csv('../datasets/files_generated/Personality/combined_features_data_out_mahalanobis_transformedDistributions.csv')
df= df.select_dtypes(['float64'])

df = df.loc[:, df.var() != 0.0]
# Create correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper_traingle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.95
to_drop_cols = [column for column in upper_traingle.columns if any(upper_traingle[column] >= 0.80)]

# Drop features 
df = df.drop(df[to_drop_cols], axis=1)
print("current shape:",df.shape)

# Determines the correlation between the variables.
fig, ax =  plt.subplots(figsize=(25,25)) 
sns.set()
ax = sns.heatmap(df.iloc[:,:30].corr(), annot=True, linewidths = 0.5, cmap='RdBu_r')
```
