```python
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
```

```python
df= pd.read_csv('../datasets/files_generated/study1_cleandata_out_mahalanobis.csv')
df=df[df['App']=='Spell']
```

```python
df=df.sort_values(by='Cond',ascending=True)
g = sns.FacetGrid(df, col="Cond") 
# g.map(sns.jointplot, "swipe_length_mean",'PQ',kind='kde')
g.map(sns.distplot, 'touch.duration',color='g')


g = sns.FacetGrid(df, col="Cond") 
# g.map(sns.jointplot, "swipe_length_mean",'PQ',kind='kde')
g.map(sns.distplot, 'swipe_length',color='r')


g = sns.FacetGrid(df, col="Cond") 
# g.map(sns.jointplot, "swipe_length_mean",'PQ',kind='kde')
g.map(sns.distplot, 'touchAccuracy_x',color='b')

g = sns.FacetGrid(df, col="Cond") 
# g.map(sns.jointplot, "swipe_length_mean",'PQ',kind='kde')
g.map(sns.distplot, 'time_between_touches')
```

```python
df= pd.read_csv('../datasets/files_generated/UX/study1_features_data_out_mahalanobis.csv')
df=df[df['App']=='Spell']

g = sns.FacetGrid(df, col="Cond") 
# g.map(sns.jointplot, "swipe_length_mean",'PQ',kind='kde')
g.map(sns.distplot, 'hit_rate',color='g')

g = sns.FacetGrid(df, col="Cond") 
# g.map(sns.jointplot, "swipe_length_mean",'PQ',kind='kde')
g.map(sns.distplot, 'swipe_length_mean',color='g')

g = sns.FacetGrid(df, col="Cond") 
# g.map(sns.jointplot, "swipe_length_mean",'PQ',kind='kde')
g.map(sns.distplot, 'swipe_length_median',color='g')

g = sns.FacetGrid(df, col="Cond") 
# g.map(sns.jointplot, "swipe_length_mean",'PQ',kind='kde')
g.map(sns.distplot, 'swipe_length_quantile',color='g')
```

```python
g = sns.JointGrid(x="swipe_length.x_mean", y="PQ", data=df) 
g.plot_joint(sns.regplot, order=2) 
g.plot_marginals(sns.distplot)
g.annotate(stats.pearsonr)
rsquare = lambda a, b: stats.pearsonr(a, b)[0] ** 2
g = g.annotate(rsquare, template="{stat}: {val:.2f}",
                stat="$R^2$", loc="upper left", fontsize=12)
plt.show()

g = sns.JointGrid(x="swipe_length.x_mean", y="ATT", data=df) 
g.plot_joint(sns.regplot, order=4) 
g.plot_marginals(sns.distplot)
g.annotate(stats.pearsonr)
rsquare = lambda a, b: stats.pearsonr(a, b)[0] ** 2
g = g.annotate(rsquare, template="{stat}: {val:.2f}",
                stat="$R^2$", loc="upper left", fontsize=12)
plt.show()

g = sns.JointGrid(x="swipe_length.x_mean", y="HQ", data=df) 
g.plot_joint(sns.regplot, order=2) 
g.plot_marginals(sns.distplot)
g.annotate(stats.pearsonr)
rsquare = lambda a, b: stats.pearsonr(a, b)[0] ** 2
g = g.annotate(rsquare, template="{stat}: {val:.2f}",
                stat="$R^2$", loc="upper left", fontsize=12)
plt.show()

# sns.jointplot("x_location.down_median", "PQ", data=df, kind='kde')
# sns.jointplot("swipe_length.x_mean", "PQ", data=df, kind='kde',color='gray')
# df.head(10)
```

```python
df= pd.read_csv('../datasets/files_generated/UX/study2_features_data_out_mahalanobis.csv')
df=df[df['App']=='Spell']
df=df.sort_values(by='Size',ascending=True)
correlations = df.select_dtypes(['float64','int64']).iloc[:,2:30].corr(method='pearson')
# plot correlation matrix
fig=plt.figure(figsize=(10,8))
sns.set(style='darkgrid')
sns.heatmap(correlations, linewidths = 0.5, cmap='coolwarm')
plt.title('Correlation Matrix plot with Pearson Coefficient correlation',fontsize=15)
# fig.savefig('Figures/Distributions/study2_correlationMatrix_visualization_mahalanobis.png')
```
