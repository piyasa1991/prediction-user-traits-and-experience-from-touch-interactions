# Data exploration for UX ratings

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# plt.style.use('ggplot')
import matplotlib.style as style
style.use('ggplot')
```

```python
df_uxratings = pd.read_csv("../datasets/rawData.firstTry/UXRatings_study1.csv",index_col=0)
df_uxratings_2 = pd.read_csv("../datasets/rawData.firstTry/UXRatings_study2.csv",index_col=0)
df_uxratings = df_uxratings[df_uxratings['App']=='Spell']
df_uxratings=df_uxratings.sort_values(by='Cond',ascending=False)
df_uxratings_2=df_uxratings_2.sort_values(by='IconSize',ascending=False)
```

```python
#ratings are given between 2.50 to 6
# average being 4.75
# lowest is 6  and 3.25
fig,ax=plt.subplots(2,2,figsize=(15,4),sharex=True)
# plt.subplot(number_of_rows + 1,4,i+1) 
sns.boxplot(x='PQ',y="Cond",data=df_uxratings,ax=ax[0][0])
sns.boxplot(x='ATT',y="Cond",data=df_uxratings,ax=ax[0][1])
sns.boxplot(x='PQ',y="IconSize",data=df_uxratings_2,ax=ax[1][0])
sns.boxplot(x='ATT',y="IconSize",data=df_uxratings_2,ax=ax[1][1])
# ax1.set_title('PQ')
ax[0][0].set_ylabel('Condition')
ax[1][0].set_ylabel('Condition')
ax[0][1].set_ylabel('')
ax[0][1].set_title('STUDY1')
# ax[0][1].set_ytickslabel('')
ax[1][1].set_ylabel('')
ax[0][0].set_xlabel('')
ax[0][1].set_xlabel('')

# fig.savefig('Figures/UX_analysis_Study1_Study2_PQ_ATT.png')
fig,ax1=plt.subplots(1,1,figsize=(20,4))
sns.distplot(df_uxratings['PQ'],kde=True,hist=False,label='PQ',ax=ax1) 
sns.distplot(df_uxratings['ATT'],kde=True,hist=False,label='ATT',ax=ax1)
# sns.distplot(df_uxratings['HQ'],kde=True,hist=False,label='HQ',ax=ax1)
```
