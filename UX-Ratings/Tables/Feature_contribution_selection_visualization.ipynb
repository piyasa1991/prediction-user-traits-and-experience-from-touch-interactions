```python
import pandas as pd
import matplotlib.pyplot as plt
import collections
```

```python
pq_study1=['swipe_length', 'x_location.release', 'touchAccuracy_x', 'difference.touch_buttonCenter_x',
           'x_location.down', 'touch.duration', 'target_touch_x_location']
pq_study2= ['touchAccuracy_x', 'touchAccuracy_y', 'difference.touch_buttonCenter_y', 'touchAccuracy']
pq_combined = ['difference.touch_buttonCenter_y', 'swipe_length.y', 
 'swipe_length.x', 'button_touch_x_location',
 'time_between_touches', 'difference.touch_buttonCenter_x']


att_study1=['swipe_length', 'touch.duration', 'target_touch_x_location', 'touchAccuracy_x']
att_study2=['touchAccuracy_x', 'button_touch_x_location', 'y_location.down', 'touchAccuracy']
att_combined=['button_touch_y_location', 'difference.touch_buttonCenter_x']
```

```python
features_set=[]
for ele in pq_study1:
    features_set.append(ele)
for ele in pq_study2:
    features_set.append(ele)
for ele in pq_combined:
    features_set.append(ele)
    
# for ele in att_study1:
#     features_set.append(ele)
# for ele in att_study2:
#     features_set.append(ele)
# for ele in att_combined:
#     features_set.append(ele)


for i,_ in enumerate(features_set):
    if features_set[i]=='touchAccuracy_x' or features_set[i]=='touchAccuracy_y':
        features_set[i]='touchAccuracy'
    if features_set[i]=='difference.touch_buttonCenter_x' or features_set[i]=='difference.touch_buttonCenter_y':
        features_set[i]='difference.touch_buttonCenter'
    if features_set[i]=='swipe_length.x' or features_set[i]=='swipe_length.y':
        features_set[i]='swipe_length'

df = pd.DataFrame.from_dict(collections.Counter(features_set),
                       orient='index',columns=['Count']).sort_values(by='Count',ascending=False)

```

```python
import seaborn as sns
plt.style.use('ggplot')
ax= sns.barplot(x='Count',y= df.index, data=df)
ax.set_xlabel('selection count')
plt.show()
```
